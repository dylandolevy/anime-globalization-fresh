```python
# 5_sentiment_prep.ipynb cell â€” aggregate ratings
import pandas as pd
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
RAW = ROOT / "data" / "raw"
PROC = ROOT / "data" / "processed"

ratings = pd.read_csv(RAW / "rating.csv", low_memory=False)
print("Ratings shape:", ratings.shape)
# common columns in Kaggle rating.csv: user_id, anime_id, rating
if "anime_id" in ratings.columns and "rating" in ratings.columns:
    agg = ratings.groupby("anime_id")["rating"].agg(["mean","count"]).reset_index().rename(columns={"mean":"rating_mean","count":"rating_count"})
    agg.to_csv(PROC / "rating_agg_by_anime_id.csv", index=False)
    print("Saved rating aggregation")
else:
    print("rating.csv format unexpected; inspect columns:", ratings.columns.tolist())
```
